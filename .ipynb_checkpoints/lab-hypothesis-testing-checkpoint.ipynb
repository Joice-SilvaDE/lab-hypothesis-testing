- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#Set the hypothesis

#H0: mu_type dragon = mu_price grass
#H1: mu_price grass != mu_price dragon

#significance level = 0.05
```

```python
df_dragon = df[df["Type 1"]== 'Dragon']["HP"] 
df_grass = df[df["Type 1"]== 'Grass']["HP"]
```

```python
st.ttest_ind(df_dragon,df_grass, equal_var=False)
```

```python
alpha = 0.05
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df
```

```python
legendary = df[df["Legendary"] == True ]
```

```python
non_legendary = df[df["Legendary"] == False ]
```

```python
types=['HP', 'Attack','Defense','Sp. Atk','Sp. Def','Speed']
for x in types:
    _ , pvalue = st.ttest_ind(legendary[x], non_legendary[x], alternative = 'two-sided')
    print(x)
    print(pvalue)
    
```
